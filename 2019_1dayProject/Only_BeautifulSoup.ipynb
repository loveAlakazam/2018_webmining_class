```python
import requests
from bs4 import BeautifulSoup

# 가져올 홈페이지 url
# Dining Code (안서동)
url='https://www.diningcode.com/list.php?query=%EC%95%88%EC%84%9C%EB%8F%99'

# request를 이용해서 url의 html 파일을 가져온다.
html = requests.get(url) 

#가져온 html 파일을 html parser를 통해서 정리한다.
soup = BeautifulSoup(html.text, "html.parser") 

print(soup)
```

```python
name_list=soup.findAll("span",attrs={"class":"btxt"}) # ' '(스페이스바) 다음부터 가게이름시작점.
address_list=soup.findAll("span",attrs={"class":"ctxt"}) #홀수만
menu_list=soup.findAll("span",attrs={"class":"stxt"})
score_list=soup.findAll("span",attrs={"class":"point"})

print('len(name_list)={}\n'.format(len(name_list)))
print('len(address_list)={}\n'.format(len(address_list)))
print('len(menu_list)={}\n'.format(len(menu_list)))
print('len(score_list)={}\n'.format(len(score_list)))


print('[ 이름 출력 ]')
for name in name_list:
    name2=name.get_text()
    print(name2[name2.index(' ')+1:])

print('\n[ 주소 출력 ]')
for address in address_list:
    idx=address_list.index(address)
    #print(idx)
    if idx%2==1:
        print(address.get_text())

print('\n[ 메뉴 출력 ]')
for menu in menu_list:
    print(menu.get_text())

print('\n[ 점수 출력 ]')
for score in score_list:
    print(score.get_text())


```

```python
Names, Address, Menu, Score=[],[],[],[]
for name in name_list:
    name2=name.get_text()
    Names.append(name2[name2.index(' ')+1:])

for address in address_list:
    idx=address_list.index(address)
    #print(idx)
    if idx%2==1:
        Address.append(address.get_text())

for menu in menu_list:
    Menu.append(menu.get_text())

for score in score_list:
    Score.append(score.get_text())
    
# Names, Address,Menu, Score 리스트 길이가 모두 같다고 할때..

store_info=[]
for i in range(len(Names)):
    store_info.append([ (i+1),Names[i], Address[i], Menu[i], Score[i]])

for i in range(len(Names)):
    print(store_info[i])
```

```python
# 그러나 100개의 데이터가 있음에도 불구하고 볼 수 있는건 10개밖에 안된다.
# 이는 더보기 버튼을 눌러야 그 이상의 정보를 볼 수있는 자바스크립트에 의한 '동적웹' 구조이기때문이다.
# BeautifulSoup 만으로는 동적웹 구조를 갖는 홈페이지에서 파싱이 불가능하다.
```
